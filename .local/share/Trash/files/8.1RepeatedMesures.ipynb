```python
# nbi:hide_in 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, interactive_output, HBox, VBox
import ipywidgets as widgets
import random
from random import choices
from scipy import stats
import scipy.stats
from IPython.display import HTML
from scipy.stats import norm
import math

groupsize=10

groupA = np.random.normal(50,10,groupsize)
groupB = groupA + np.random.normal(-5,.5,len(groupA))  #the expected value of diffrence is ALSO 10
groupB = np.random.choice(groupB,size=len(groupB),replace=False)
groupC = groupA + np.random.normal(0,0,len(groupA))  #the expected value of diffrence is ALSO 10
groupC = np.random.choice(groupC,size=len(groupC),replace=False)

plt.figure(figsize=(14,8))

for i in range(len(groupA)):
  plt.plot([0,1],[groupA[i],groupB[i]],color="grey",linestyle='--', linewidth = 3)
for i in range(len(groupB)):
  plt.plot([1,2],[groupB[i],groupC[i]],color="grey",linestyle='--', linewidth = 3)

plt.scatter(np.repeat("Incongruent",len(groupA)), groupA,color = "silver", s = 200)
plt.scatter(np.repeat("Neutral",len(groupB)), groupB,color = "tan", s = 200)
plt.scatter(np.repeat("Congruent",len(groupC)), groupC,color = "teal", s = 200)

plt.yticks(fontsize = 18, color = "black")
plt.xticks(fontsize = 22, color = "black")


plt.title("Stroop Effect per Participant", fontsize = 36);

```
